```
#argument 1: location of CSV file
flights = spark.read.option("header","true").csv("dbfs:/mnt/ktbigdatapart1/995682566_T_T100D_SEGMENT_US_CARRIER_ONLY.csv")

#argument 2: maximum number of iteration
iteration = 10

#argument 3: location of output file
outputfile = "dbfs:/mnt/ktbigdatapart1/outputpart1"

flights.show()
```

```
airports = flights.select("ORIGIN_AIRPORT_ID", "ORIGIN").toDF("id","name").distinct()

airports.show()
```

```
airportEdges = flights.select("ORIGIN", "DEST").toDF("src","dst")

airportEdges.show()
```

```
#getting list of src and destination airport
from pyspark.sql.functions import concat, concat_ws, lit, col, trim
edges = airportEdges.select(concat(col("src"),lit(" "), col("dst"))).rdd.flatMap(lambda x: x)

edges.collect()
```

```
#reduce by key to combine list of airports and their outgoing list
#now we have key and value pair: (airport_i, [outgoing list from aiport_i])
edgeRdd = edges.map(lambda x:tuple(x.split(" "))).map(lambda x:(x[0],[x[1]])).reduceByKey(lambda x, y: x+y)

edgeRdd.collect()
```

```
#airportGraph.edges.where("src = 'ROC'").show()
```

```
##initialize all airport's page rank to 10.0
pageRank = edgeRdd.map(lambda x: (x[0], 10.0))

pageRank.collect()
```

```
#compute pageRank_i/outDegree_i
def compute(airports, rank):
  for airport in airports:
    yield(airport, rank/len(airports))
```

```
from graphframes import GraphFrame

airportGraph = GraphFrame(airports, airportEdges)

airportGraph.cache()

outDeg = airportGraph.outDegrees.toDF("id", "outDegree")
N = outDeg.count()
```

```
#for each iteration, combine the pageRank_i/outDegree_i sum and add/multiply alpha
for i in range(iteration):
  pageRank = pageRank.join(edgeRdd).flatMap(lambda x : compute(x[1][1], x[1][0])).reduceByKey(lambda x,y: x+y).mapValues(lambda x: 0.85*x + 0.15/N)

pageRank.collect()
```

```
#output to DataFrame and sort result by desc order
from pyspark.sql.functions import desc

resultDF = spark.createDataFrame(pageRank).toDF("Airport", "PageRank")

resultDF.orderBy(desc("PageRank")).show()
```

```
#write output DataFrame to file
resultDF = resultDF.orderBy(desc("PageRank"))
resultDF.write.format("csv").mode("overwrite").save(outputfile)
```
